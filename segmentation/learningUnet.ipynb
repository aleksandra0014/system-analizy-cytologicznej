## Imports

```python
import torch
from dataset import CellNucleusDataset, train_loader, val_loader
from modelsUnet import UNet, train, UNet4Levels
import torch.nn as nn
from torch.utils.data import DataLoader
import os
```

## Parameters

```python
batch_size = 4
learning_rate = 1e-4
num_epochs = 50
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

## Training

```python
model = UNet(in_channels=3, out_channels=2)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

train(model, train_loader, val_loader, optimizer, device, epochs=num_epochs, log_dir="runs/unet_cell_nucleus/0908", model_path="unet_cell_nucleus_4_50_1410.pth")

torch.save(model.state_dict(), "unet_cell_nucleus_4_50_1410.pth")
```

```python
model = UNet4Levels(in_channels=3, out_channels=2)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

train(model, train_loader, val_loader, optimizer, device, epochs=num_epochs, log_dir="runs/unet_cell_nucleus/0908", model_path="unet4_cell_nucleus_4_50_1310.pth")

torch.save(model.state_dict(), "unet4_cell_nucleus_4_50_1310.pth")
```

```python
model = UNet(in_channels=3, out_channels=2)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

train(model, train_loader, val_loader, optimizer, device, epochs=50, log_dir="runs/unet_cell_nucleus/610", model_path="unet_cell_nucleus_1310_50.pth")

torch.save(model.state_dict(), "unet_cell_nucleus_1310_50.pth")
```

## Importy

```python
from dataset import train_dataset, test_dataset, val_dataset, class_counts
from torch.utils.data import DataLoader
from models import CytologyClassifier
```

## Parametry

```python
batch_size = 24
lr = 0.001
num_epochs = 50
```

## Podział na loadery

```python
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
test_loader = DataLoader(test_dataset, batch_size=batch_size)
```

## Nauka modeli

## SVM

```python

import numpy as np
import torch
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from models import SVMClassifier
```

```python


# Load trained ResNet model
model = CytologyClassifier(num_classes=3, architecture='resnet18')
model.load(r'C:\Users\aleks\OneDrive\Documents\inzynierka\classification_models\resnet18\32_0.0001_50_06.07.pth')
model.model.eval()
svm = SVMClassifier(model=model, num_classes=3, model_name='resnet18')

features, labels = svm.get_features(model=model, data_loader=train_loader)
svm.fit(features, labels)

test_features, test_labels = svm.get_features(model=model, data_loader=test_loader)
svm_preds = svm.predict(test_features)
svm.evaluate(model, svm_preds, test_labels)
```

```python
import joblib


joblib.dump(svm.svm, 'svm_resnet18.pkl')
```

```python
model = CytologyClassifier(num_classes=3, architecture='vgg16')
model.load(r'C:\Users\aleks\OneDrive\Documents\inzynierka\classification_models\vgg16\24_0.001_50_07.07.pth')
model.model.eval()
svm = SVMClassifier(model=model, model_name='vgg16', num_classes=3)

features, labels = svm.get_features(model=model, model_name='vgg16', data_loader=train_loader)
svm.fit(features, labels)

test_features, test_labels = svm.get_features(model=model, model_name='vgg16', data_loader=test_loader)
svm_preds = svm.predict(test_features)
svm.evaluate(model, svm_preds, test_labels)
```

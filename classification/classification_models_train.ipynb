## Importy

```python
from dataset import train_dataset, test_dataset, val_dataset, class_counts
from torch.utils.data import DataLoader
from models import CytologyClassifier
from models_testing import test_model
from dataset import test_dataset
from torch.utils.data import DataLoader
```

## Parametry

```python
batch_size = 16
lr = 0.001
num_epochs = 50
```

## Podział na loadery

```python
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
test_loader = DataLoader(test_dataset, batch_size=batch_size)
```

## Nauka modeli

```python
classifier = CytologyClassifier(num_classes=3, lr=0.001, architecture='custom_cnn', class_counts=class_counts)
classifier.train(train_loader=train_loader, val_loader=val_loader, num_epochs=50, save_best_path=r'classification_models\custom_cnn\16_0_001_50_0908.pth')
```

```python
test_loader = DataLoader(test_dataset)
test_model(r'classification_models\custom_cnn\16_0_001_50_0908.pth', test_loader, architecture='custom_cnn', model_label='custom_cnn')
```

```python

import numpy as np
import torch
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from models import SVMClassifier
```

```python


# Load trained ResNet model
model = CytologyClassifier(num_classes=3, architecture='resnet18')
model.load(r'C:\Users\aleks\OneDrive\Documents\inzynierka\classification_models\resnet18\32_0.0001_50_06.07.pth')
model.model.eval()
svm = SVMClassifier(model=model, num_classes=3, model_name='resnet18')

features, labels = svm.get_features(model=model, data_loader=train_loader)
svm.fit(features, labels)

test_features, test_labels = svm.get_features(model=model, data_loader=test_loader)
svm_preds = svm.predict(test_features)
svm.evaluate(model, svm_preds, test_labels)
```

```python
import joblib


joblib.dump(svm.svm, 'svm_resnet18.pkl')
```

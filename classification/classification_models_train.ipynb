```python
from dataset import train_dataset, test_dataset, val_dataset, class_counts
from torch.utils.data import DataLoader
from models import CytologyClassifier
from models_testing import test_model
from dataset import test_dataset
from torch.utils.data import DataLoader
```

## Parametry

```python
batch_size = 32
lr = 0.001
num_epochs = 50
```

## Podział na loadery

```python
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
test_loader = DataLoader(test_dataset, batch_size=batch_size)
```

```python
import os

def prepare_dataset_from_folders(root_dir, class_names):
    image_paths = []
    labels = []

    for idx, class_name in enumerate(class_names):
        class_folder = os.path.join(root_dir, class_name)
        for filename in os.listdir(class_folder):
            if filename.lower().endswith(('.jpg', '.png', '.bmp')):
                image_paths.append(os.path.join(class_folder, filename))
                labels.append(idx)

    return image_paths, labels

# Przygotuj dane
data_dir = r'C:\Users\aleks\OneDrive\Documents\inzynierka\data\data_single_cropped2'
class_names = ['HSIL', 'LSIL', 'NSIL']
image_paths, labels = prepare_dataset_from_folders(data_dir, class_names)
```

```python
from torchvision import datasets, transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5]*3, std=[0.5]*3)
])

classifier = CytologyClassifier(num_classes=3, lr=lr, architecture='custom_cnn', class_counts=class_counts)
classifier.cross_validate(image_paths=image_paths, labels=labels, transform=transform, num_epochs=num_epochs, k_folds=5)
```

```python
classifier = CytologyClassifier(num_classes=3, lr=lr, architecture='vgg16', class_counts=class_counts)
classifier.cross_validate(image_paths=image_paths, labels=labels, transform=transform, num_epochs=num_epochs, k_folds=5)
```

```python
classifier = CytologyClassifier(num_classes=3, lr=lr, architecture='resnet18', class_counts=class_counts)
classifier.cross_validate(image_paths=image_paths, labels=labels, transform=transform, num_epochs=num_epochs, k_folds=5)
```

## Nauka modeli

```python
classifier = CytologyClassifier(num_classes=3, lr=lr, architecture='custom_cnn', class_counts=class_counts)
classifier.train(train_loader=train_loader, val_loader=val_loader, num_epochs=num_epochs, save_best_path=r'classification_models\custom_cnn\24_0_001_50_2407.pth')
```
